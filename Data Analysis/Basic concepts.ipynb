### Part 1: NumPy — The Foundation of Data Analysis
NumPy is the backbone of scientific computing in Python.  
It allows you to work with arrays, matrices, and perform fast numerical operations.

#### Step 1.1 — Creating Arrays

```python
import numpy as np

# Create a 1D array
arr = np.array([10, 20, 30, 40, 50])
print("1D array = ",arr)

# Create a 2D array (matrix)
matrix = np.array([[1, 2, 3],[4, 5, 6]])
print("\n2D array = ",matrix)
```

#### Step 1.2 — Array Properties

```python
print("Shape:", matrix.shape)      # (2, 3)
print("\nDimensions:", matrix.ndim)  # 2D
print("\nData type:", matrix.dtype)  # int64 or int32
print("\nSize:", matrix.size)        # total elements
```

#### Step 1.3 — Range and Random Arrays

```python
# Using arange
a = np.arange(0, 10, 2)  # Start=0, Stop=10, Step=2
print("Using arange",a)

# Using linspace
b = np.linspace(0, 1, 5)  # 5 equally spaced points between 0 and 1
print("\nUsing linspace",b)

# Random numbers
r = np.random.rand(3, 2)  # 3x2 matrix of random floats (0 to 1)
print("\nRandom numbers",r)
```

#### Step 1.4 — Indexing and Slicing

```python
arr = np.array([10, 20, 30, 40, 50])

print(arr[0])   # First element
print("\n",arr[-1])  # Last element
print("\n",arr[1:4]) # Slice 2nd to 4th element
```

#### Step 1.5 — Mathematical Operations

```python
arr = np.array([1, 2, 3, 4, 5])

print(arr + 10)  # Add 10 to every element
print("\n",arr * 2)   # Multiply each element by 2
print("\n",np.sqrt(arr))  # Square root
print("\n",np.exp(arr))   # Exponential
print("\n",np.mean(arr))  # Mean value
```

#### Step 1.6 — Matrix Operations

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("\n Add=",A + B)        # Element-wise addition
print("\n Multiple=",A * B)        # Element-wise multiplication
print("\n Matix multiple=",A @ B)        # Matrix multiplication
print("\n",np.dot(A, B)) # Same as @
```

#### Step 1.7 — Statistics and Aggregation

```python
data = np.random.randint(10, 100, (5, 4))  # 5x4 matrix
print(data)
print("\n")
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Sum by column:", np.sum(data, axis=0))
print("Max by row:", np.max(data, axis=1))
```

### Part 2: Pandas — DataFrames and Real-World
Pandas builds on NumPy and provides labels (columns & indices), making data analysis more readable.

#### Step 2.1 — Import and Create DataFrame

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Age': [24, 27, 22, 32],
    'Salary': [50000, 54000, 58000, 62000]
}

df = pd.DataFrame(data)
print(df)
```

#### Step 2.2 — Explore the Data

```python
print(df.head())    # First 5 rows
```

```python
print(df.tail())    # Last 5 rows
```

```python
print(df.info())    # Data types and non-null counts
```

```python
print(df.describe())# Statistics summary
```

#### Step 2.3 — Selecting Data

```python
print(df['Name'])          # Single column (Series)

print("\n")

print(df[['Name', 'Age']]) # Multiple columns
```

```python
print(df.iloc[0])          # First row by index position

print("\n")

print(df.loc[0, 'Name'])   # Access specific cell
```

#### Step 2.4 — Filtering Data

```python
print(df[df['Age'] > 25])  # Filter rows

print("\n")

print(df[(df['Age'] > 25) & (df['Salary'] > 52000)])
```

#### Step 2.5 — Add, Update, and Drop Columns

```python
df['Bonus'] = df['Salary'] * 0.1
print(df)
```

```python
df['Dept'] = ['IT', 'HR', 'IT', 'Finance']
print(df)
```

```python
df.drop('Bonus', axis=1, inplace=True)
print(df)
```

#### step 2.6 — Handling Missing Data

```python
df.loc[2, 'Age'] = np.nan  # Simulate missing data
print(df)
```

```python
df.fillna(df['Age'].mean()) # try -- dropna() removes them
print(df)
```

#### Step 2.7 — Grouping and Aggregation

```python
dept_salary = df.groupby('Dept')['Salary'].mean()
print(dept_salary)
```

#### Step 2.8 — Merging and Joining

```python
dept_info = pd.DataFrame({
    'Dept': ['IT', 'HR', 'Finance'],
    'Manager': ['John', 'Sophie', 'Mark']
})

merged = pd.merge(df, dept_info, on='Dept', how='left')
print(merged)
```

### Part 3: Matplotlib — Core Data Visualization
Matplotlib is the foundation library for creating static, publication-quality plots in Python.

#### Step 3.3 — Basic Line Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 16]

plt.figure(figsize=(8,4))
plt.plot(x, y, color='blue', marker='o', linestyle='--', linewidth=2)
plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

#### Step 3.2 — Bar Chart

```python
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 15, 30]

plt.bar(categories, values)
plt.title('Bar Chart Example')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

#### Step 3.3 — Histogram (Data Distribution)

```python
import numpy as np

data = np.random.randn(1000)  # 1000 random values

plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

#### Step 3.4 — Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x, y, c=colors, s=80, alpha=0.7)
plt.title('Scatter Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

#### Step 3.5 — Subplots

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].plot(x, y, 'o-')
axs[0].set_title('Left Plot')

axs[1].hist(data, bins=20)
axs[1].set_title('Right Plot')

plt.tight_layout()
plt.show()
```

### Part 4: Seaborn — Statistical Data Visualization
Seaborn is built on Matplotlib but gives you cleaner, more beautiful plots with one-liner commands.

#### Step 4.1 — Import and Sample Data

```python
import seaborn as sns

# Built-in dataset
df = sns.load_dataset('tips')
df.head()
```

#### Step 4.2 — Scatterplot with Regression Line

```python
sns.lmplot(data=df, x='total_bill', y='tip', hue='sex', height=5, aspect=1.5)
```

#### Step 4.3 — Histogram and KDE (Kernel Density Estimation) Plot

```python
sns.histplot(df['total_bill'], bins=20, kde=True)
plt.title('Distribution of Total Bill')
plt.show()
```

#### Step 4.4 — Box Plot

```python
sns.boxplot(data=df, x='day', y='total_bill', hue='sex')
plt.title('Boxplot of Total Bill by Day and Gender')
plt.show()
```

#### Step 4.5 — Pairplot

```python
sns.pairplot(df, hue='sex')
plt.show()
```

#### Step 4.6 — Correlation Heatmap

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

### Part 5: Data Cleaning & Preprocessing

```python
import pandas as pd
import numpy as np

# Sample raw data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Evan', 'Fiona'],
    'Age': [25, np.nan, 35, 28, 40, None],
    'Gender': ['Female', 'Male', 'Male', 'Female', np.nan, 'Female'],
    'Department': ['IT', 'HR', 'IT', np.nan, 'Finance', 'Finance'],
    'Salary': [50000, 54000, None, 58000, 62000, 62000],
    'Experience(Years)': [2, 5, 10, 3, None, 4]
}

df = pd.DataFrame(data)
df
```

```python
pd.set_option('future.no_silent_downcasting', True)
```

```python
df.isnull().sum()  # Detect Missing Values
```

```python
# Fill numeric columns with mean
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Experience(Years)']=df['Experience(Years)'].fillna(df['Experience(Years)'].median())
df['Salary']=df['Salary'].fillna(df['Salary'].mean())
```

```python
# Fill Gender safely
if not df['Gender'].mode().empty:
    df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
else:
    df['Gender'] = df['Gender'].fillna('Unknown')

# Fill Department safely
if not df['Department'].mode().empty:
    df['Department'] = df['Department'].fillna(df['Department'].mode()[0])
else:
    df['Department'] = df['Department'].fillna('Unknown')
```

```python
df.isnull().sum()  # Detect Missing Values
```

#### Step 6.4 — Detecting and Removing Duplicates

```python
# Add a duplicate row to demonstrate
df.loc[6] = df.loc[5]

print("Before removing duplicates:", len(df))
df.drop_duplicates(inplace=True)
print("After removing duplicates:", len(df))
```
