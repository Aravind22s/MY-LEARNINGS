````python
from google import genai
import os
from dotenv import load_dotenv

load_dotenv()

client = genai.Client(api_key=os.getenv("Gemini_api_key"))
````

````python
def get_response (prompt, model="gemini-3.1-flash-lite"):
    response = client.models.generate_content(
        model = model,
        contents = prompt,
    )
    
    return response.text
````

#### Delimiters are special characters or markers used to clearly separate different parts of a prompt. They help the AI distinguish between instructions, input data, examples, and expected output.

### Triple Quotes (""" """) & Custom Delimiters & Variable + Delimiters

Used for long text, articles, documents

````python
article = """Artificial Intelligence is transforming healthcare by helping doctors diagnose diseases faster."""

prompt1 = f"""
Summarize the following article.
<<<START>>> 
{article}
<<<END>>>
"""

print(get_response(prompt1))
````

### Triple Backticks (```)

Use: Code explanation

````python
prompt2 = """
Explain the following Python code.

```python
def add(a, b):
    return a + b

print(add(10, 20))
```
"""

print(get_response(prompt2))
````

### XML Delimiters

Use: Multiple sections

````python
prompt3 = """
<role>
You are an experienced Python teacher.
</role>

<task>
Explain Python functions.
</task>

<audience>
First-year college students.
</audience>
"""

print(get_response(prompt3))
````

### Markdown Headings

Use: Professional prompts

````python
prompt4 = """
# Role
You are a Data Scientist.

# Task
Explain Linear Regression.

# Audience
Beginner students.

# Output
Use simple language and examples.
"""

print(get_response(prompt4))
````

### JSON Delimiter

Use: APIs and structured data

````python
prompt5 = """
Analyze this JSON data.

{
    "name": "Aravind",
    "age": 22,
    "skills": [
        "Python",
        "Machine Learning",
        "AI"
    ]
}
"""

print(get_response(prompt5))
````
