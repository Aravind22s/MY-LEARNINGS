# SQL (Structured Query Language)

**SQL databases** are relational database management systems (RDBMS) that store data in structured tables consisting of rows and columns. They use SQL to define, manipulate, and query data, making them ideal for applications requiring strong consistency and complex relationships.

### SQLite
**SQLite** is a lightweight, serverless, self-contained relational database management system (RDBMS) that stores the entire database in a single file. It is widely used in desktop, mobile, and embedded applications because it requires no separate database server and is easy to set up.

| Method              | Description                                          |
| ------------------- | ---------------------------------------------------- |
| `sqlite3.connect()` | Connects to the database                             |
| `cursor()`          | Creates a cursor object                              |
| `execute()`         | Executes a single SQL statement                      |
| `executemany()`     | Executes the same SQL statement for multiple records |
| `fetchone()`        | Retrieves one row                                    |
| `fetchmany(n)`      | Retrieves the next `n` rows                          |
| `fetchall()`        | Retrieves all rows                                   |
| `commit()`          | Saves changes to the database                        |
| `rollback()`        | Reverts uncommitted changes                          |
| `close()`           | Closes the database connection                       |


```python
import sqlite3

# Connect to database (creates the database if it doesn't exist)
conn = sqlite3.connect("student.db")
print("Database connected successfully!")
# Close the connection
conn.close()

print("Database connection closed.")
```

```python
import sqlite3

# Connect to database
conn = sqlite3.connect("student.db")
# Create cursor object
cursor = conn.cursor()
# Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    student_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    age INTEGER,
    department TEXT,
    cgpa REAL
)
""")
# Save changes
conn.commit()
print("Table created successfully!")

# Close connection
conn.close()
```

```python
import sqlite3

conn = sqlite3.connect("student.db")
cursor = conn.cursor()

students = [
    ("Rahul", 21, "CSE", 8.5),
    ("Priya", 22, "ECE", 9.1),
    ("Kumar", 20, "IT", 8.0),
    ("Meena", 21, "AI & DS", 9.3)
]

cursor.executemany("""
INSERT INTO students (name, age, department, cgpa)
VALUES (?, ?, ?, ?)
""", students)

conn.commit()
print("Multiple records inserted!")

conn.close()
```

```python
import sqlite3

conn = sqlite3.connect("student.db")
cursor = conn.cursor()

cursor.execute("SELECT * FROM students")
rows = cursor.fetchall()

for row in rows:
    print(row)

conn.close()
```

```python
import sqlite3

conn = sqlite3.connect("student.db")
cursor = conn.cursor()

department = "AI & DS"

cursor.execute("""
SELECT * FROM students
WHERE department = ?
""", (department,))

rows = cursor.fetchall()

for row in rows:
    print(row)

conn.close()
```

```python
# Using Context Manager (Recommended)
# Using with automatically commits (if no error) and closes the connection.

import sqlite3

with sqlite3.connect("student.db") as conn:
    cursor = conn.cursor()

    cursor.execute("""
    SELECT * FROM students
    """)

    rows = cursor.fetchall()

    for row in rows:
        print(row)
```

```python
import sqlite3

DB_NAME = "student.db"

# Connect
conn = sqlite3.connect(DB_NAME)
cursor = conn.cursor()

# Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    age INTEGER,
    department TEXT,
    cgpa REAL
)
""")

# Insert
cursor.execute("""
INSERT INTO students(name, age, department, cgpa)
VALUES (?, ?, ?, ?)
""", ("Aravind", 22, "AI & DS", 9.2))

# Read
cursor.execute("SELECT * FROM students")
print(cursor.fetchall())

# Update
cursor.execute("""
UPDATE students
SET cgpa = ?
WHERE name = ?
""", (9.5, "Aravind"))

# Delete
cursor.execute("""
DELETE FROM students
WHERE name = ?
""", ("Aravind",))

# Save changes
conn.commit()

# Close
conn.close()
```

# NoSQL (Not Only SQL)

**NoSQL databases** are non-relational databases designed to store semi-structured, unstructured, or rapidly changing data using models such as document, key-value, column-family, or graph. They are highly scalable, flexible, and well-suited for big data, real-time applications, and distributed systems.


### MongoDB

**MongoDB** is a document-oriented NoSQL database that stores data as flexible, JSON-like BSON documents instead of tables and rows. It is designed for scalability, high performance, and applications with dynamic or unstructured data.

| Method / Command                 | Description                                                        |
| -------------------------------- | ------------------------------------------------------------------ |
| `show dbs`                       | Displays all available databases.                                  |
| `use database_name`              | Creates or switches to the specified database.                     |
| `db`                             | Displays the currently selected database.                          |
| `db.dropDatabase()`              | Deletes the current database.                                      |
| `show collections`               | Lists all collections in the current database.                     |
| `db.createCollection("name")`    | Creates a new collection.                                          |
| `db.collection.drop()`           | Deletes a collection.                                              |
| `db.collection.insertOne()`      | Inserts a single document into a collection.                       |
| `db.collection.insertMany()`     | Inserts multiple documents into a collection.                      |
| `db.collection.find()`           | Retrieves all matching documents from a collection.                |
| `db.collection.findOne()`        | Retrieves the first matching document.                             |
| `db.collection.updateOne()`      | Updates the first document that matches the filter.                |
| `db.collection.updateMany()`     | Updates all documents that match the filter.                       |
| `db.collection.replaceOne()`     | Replaces an entire document with a new document.                   |
| `db.collection.deleteOne()`      | Deletes the first matching document.                               |
| `db.collection.deleteMany()`     | Deletes all matching documents.                                    |
| `db.collection.countDocuments()` | Returns the total number of matching documents.                    |
| `db.collection.find().sort()`    | Sorts query results in ascending or descending order.              |
| `db.collection.find().limit()`   | Limits the number of documents returned.                           |
| `db.collection.find().skip()`    | Skips a specified number of documents.                             |
| `db.collection.aggregate()`      | Performs aggregation operations such as grouping and calculations. |


```python
# pip install pymongo

from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/") # connection string

# show dbs
print("Databases:")
print(client.list_database_names())

# use database_name
db = client["student_db"]

print("\nCurrent Database:")
print(db.name)

# db.createCollection("students")
if "students" not in db.list_collection_names():
    db.create_collection("students")
    print("\nCollection Created")
else:
    print("\nCollection Already Exists")

# show collections
print("\nCollections:")
print(db.list_collection_names())

# Access Collection
collection = db["students"]
```

```python
# insertOne()
collection.insert_one({
    "name": "Aravind",
    "age": 22,
    "department": "AI & DS",
    "cgpa": 9.3
})
print("\nOne Document Inserted")

# insertMany()
collection.insert_many([
    {
        "name": "Rahul",
        "age": 21,
        "department": "CSE",
        "cgpa": 8.5
    },
    {
        "name": "Priya",
        "age": 20,
        "department": "ECE",
        "cgpa": 9.1
    },
    {
        "name": "Kumar",
        "age": 23,
        "department": "IT",
        "cgpa": 8.8
    }
])
print("Multiple Documents Inserted")
```

```python
# updateOne()

collection.update_one(
    {"name": "Aravind"},
    {"$set": {"cgpa": 9.7}}
)

print("\nOne Document Updated")

# updateMany()

collection.update_many(
    {},
    {"$set": {"status": "Active"}}
)

print("Multiple Documents Updated")

# replaceOne()

collection.replace_one(
    {"name": "Rahul"},
    {
        "name": "Rahul",
        "age": 22,
        "department": "CSE",
        "cgpa": 8.9
    }
)

print("One Document Replaced")
```

```python
# find()

print("\nAll Students")

for student in collection.find():
    print(student)

# findOne()

print("\nFind One")

student = collection.find_one({"name": "Aravind"})
print(student)

# countDocuments()

count = collection.count_documents({})
print("\nTotal Documents:", count)

# find().sort()

print("\nSorted by CGPA (Descending)")

for student in collection.find().sort("cgpa", -1):
    print(student)

# find().limit()

print("\nTop 2 Students")

for student in collection.find().limit(2):
    print(student)
    
# find().skip()

print("\nSkip First 2 Students")

for student in collection.find().skip(2):
    print(student)
```

```python
# aggregate()

print("\nAverage CGPA by Department")

pipeline = [
    {
        "$group": {
            "_id": "$department",
            "Average_CGPA": {"$avg": "$cgpa"}
        }
    }
]

for result in collection.aggregate(pipeline):
    print(result)
```

```python
# deleteOne()

collection.delete_one({"name": "Kumar"})
print("\nOne Document Deleted")
```

```python
# deleteMany()

collection.delete_many({"status": "Active"})
print("Multiple Documents Deleted")
```

```python
# delete the collection

collection.drop()
print("Collection Deleted")
```

```python
#  delete the database

client.drop_database("student_db")
print("Database Deleted")
```

```python
# Close Connection

client.close()
print("\nMongoDB Connection Closed")
```

### Redis

**Redis** is an in-memory NoSQL key-value database used for ultra-fast data storage and retrieval. It is commonly used for caching, session management, real-time analytics, and message queuing due to its high-speed performance.

| Command                 | Description                                              |
| ----------------------- | -------------------------------------------------------- |
| `SET key value`         | Stores a key-value pair.                                 |
| `GET key`               | Retrieves the value of a key.                            |
| `DEL key`               | Deletes one or more keys.                                |
| `EXISTS key`            | Checks whether a key exists.                             |
| `MSET key value ...`    | Stores multiple key-value pairs at once.                 |
| `MGET key1 key2 ...`    | Retrieves values for multiple keys.                      |
| `INCR key`              | Increments the integer value of a key by 1.              |
| `DECR key`              | Decrements the integer value of a key by 1.              |
| `LPUSH key value`       | Inserts an element at the beginning of a list.           |
| `RPUSH key value`       | Inserts an element at the end of a list.                 |
| `LPOP key`              | Removes and returns the first element of a list.         |
| `RPOP key`              | Removes and returns the last element of a list.          |
| `LRANGE key start stop` | Retrieves elements from a list within a specified range. |
| `SADD key member`       | Adds one or more members to a set.                       |
| `SMEMBERS key`          | Returns all members of a set.                            |
| `SREM key member`       | Removes a member from a set.                             |
| `HSET key field value`  | Sets a field-value pair in a hash.                       |
| `HGET key field`        | Retrieves the value of a specific hash field.            |
| `HGETALL key`           | Retrieves all fields and values from a hash.             |
| `HDEL key field`        | Deletes one or more fields from a hash.                  |
| `ZADD key score member` | Adds a member with a score to a sorted set.              |
| `ZRANGE key start stop` | Retrieves members from a sorted set.                     |
| `KEYS *`                | Lists all keys in the database.                          |
| `EXPIRE key seconds`    | Sets an expiration time for a key.                       |
| `TTL key`               | Returns the remaining time to live for a key.            |


```python
# pip install redis
import redis

# Connect to Redis
r = redis.Redis(
    host="localhost",
    port=6379,
    db=0,
    decode_responses=True
)

print("Connected to Redis Successfully!")
```

```python
# SET
r.set("username", "Aravind")
print("SET Command Executed")

# GET
print("\nGET")
print(r.get("username"))

# EXISTS
print("\nEXISTS")
print(r.exists("username"))
```

```python
# MSET
r.mset({
    "city": "Chennai",
    "course": "AI & DS"
})
print("\nMSET Command Executed")

# MGET
print("\nMGET")
print(r.mget("username", "city", "course"))

# INCR
r.set("counter", 10)
r.incr("counter")
print("\nCounter:", r.get("counter"))

# DECR
r.decr("counter")
print("Counter:", r.get("counter"))
```

```python
# LPUSH
r.lpush("students", "Aravind")
print("\nLPUSH Executed")

# RPUSH
r.rpush("students", "Rahul")
r.rpush("students", "Priya")
print("RPUSH Executed")

# LRANGE
print("\nStudent List")
print(r.lrange("students", 0, -1))
```

```python
# LPOP
print("\nLPOP")
print(r.lpop("students"))

# RPOP
print("\nRPOP")
print(r.rpop("students"))
```

```python
# SADD
r.sadd("skills", "Python", "SQL", "MongoDB")
print("\nSADD Executed")

# SMEMBERS
print("\nSMEMBERS")
print(r.smembers("skills"))

# SREM
r.srem("skills", "SQL")
print("\nAfter SREM")
print(r.smembers("skills"))
```

```python
# HSET
r.hset(
    "student",
    mapping={
        "name": "Aravind",
        "age": 22,
        "department": "AI & DS"
    }
)
print("\nHash Created")

# HGET
print("\nHGET")
print(r.hget("student", "name"))

# HGETALL
print("\nHGETALL")
print(r.hgetall("student"))

# HDEL
r.hdel("student", "age")
print("\nAfter HDEL")
print(r.hgetall("student"))
```

```python
# ZADD
r.zadd(
    "marks",
    {
        "Aravind": 95,
        "Rahul": 90,
        "Priya": 98
    }
)
print("\nZADD Executed")

# ZRANGE
print("\nZRANGE")
print(r.zrange("marks", 0, -1, withscores=True))
```

```python
# KEYS
print("\nAll Keys")
print(r.keys("*"))

# EXPIRE
r.set("temp", "Hello")
r.expire("temp", 60)
print("\nExpiry Set")

# TTL
print("\nTTL")
print(r.ttl("temp"))

# DEL
r.delete("temp")
print("\nKey Deleted")
print(r.exists("temp"))
```
