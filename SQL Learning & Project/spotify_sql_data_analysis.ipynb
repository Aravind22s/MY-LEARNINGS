# Retrieve all columns top 3 entries for songs order by energy ( high value ) 
released in a 
2010, 2015, 2019 )

```python
%%sql

WITH top3_spotify_entries AS(SELECT *,ROW_NUMBER() OVER(PARTITION BY YEAR ORDER BY nrgy desc) AS rownum
FROM spotify
WHERE year in (2019,2018,2017,2016))

SELECT * FROM top3_spotify_entries 
WHERE rownum <=3;
```

average danceability (dnce)

```python
%%sql

SELECT `top genre`, AVG(dnce) AS AVG_DNCE
FROM SPOTIFY
GROUP BY `top genre`
ORDER BY AVG_DNCE DESC
LIMIT 5
```

# Find the song with the highest energy (nrgy) value for each year.

```python
%%sql

WITH highest_energy_level AS (SELECT title,year,nrgy,
ROW_NUMBER() OVER (PARTITION BY year ORDER BY nrgy DESC) AS HIGHEST_ENERGY
FROM spotify)

SELECT * FROM highest_energy_level
where HIGHEST_ENERGY<=1
```

# Calculate the median value of the "val" column

```python
%%sql

WITH SORTDATA AS (SELECT *, 
ROW_NUMBER() OVER (ORDER BY val ASC) AS ROW_NUM,
COUNT(*) OVER () AS total_count
FROM SPOTIFY)
                  
SELECT AVG(val) AS MEDIAN_VALUE
FROM SORTDATA
WHERE ROW_NUM IN (FLOOR(total_count / 2), CEIL(total_count / 2));
```

# To calculate median value of the "val" column within different categories like top genre

```python
%%sql

WITH SORTDATA AS (SELECT *, 
ROW_NUMBER() OVER (PARTITION BY `top genre` ORDER BY val ASC) AS ROW_NUM,
COUNT(*) OVER (PARTITION BY`top genre`) AS total_count
FROM SPOTIFY)
                  
SELECT `top genre`,AVG(val) AS MEDIAN_VALUE
FROM SORTDATA
WHERE ROW_NUM IN (FLOOR(total_count / 2), CEIL(total_count / 2))
GROUP BY `top genre`;
```

# Identify the artist with the most songs having a BPM greater than 120.

```python
%%sql

SELECT COUNT(*) AS NUM_SONGS,artist
FROM spotify
WHERE bpm < 120
GROUP BY artist
ORDER BY NUM_SONGS DESC
LIMIT 5;
```

# Detect any outliers in the "dur" (duration) column using statistical methods.

```python
%%sql
# Z-SCORE METHOD
WITH stats AS (
    SELECT AVG(dur) AS mean_dur, STDDEV_POP(dur) AS std_dev_dur
    FROM spotify
)
SELECT dur,
    CASE 
        WHEN ABS(dur - mean_dur) > 3 * std_dev_dur THEN 'Outlier'
        ELSE 'Not an Outlier'
    END AS outlier_status
FROM spotify, stats
HAVING outlier_status = 'Outlier';


```

# Identify any correlations between BPM and energy levels across different genres.

```python
%%sql

SELECT `top genre`,correlation,
CASE
    WHEN correlation > 0 THEN 'Positive Correlation'
    WHEN correlation < 0 THEN 'Negative Correlation'
    WHEN correlation = 0 THEN 'Zero Correlation'
END AS correlation_status
FROM (SELECT `top genre`,
        (COUNT(*) * SUM(bpm * nrgy) - SUM(bpm) * SUM(nrgy)) /
        (SQRT((COUNT(*) * SUM(bpm * bpm) - SUM(bpm) * SUM(bpm))
        * (COUNT(*) * SUM(nrgy * nrgy) - SUM(nrgy) * SUM(nrgy)))) AS correlation
FROM spotify
GROUP BY `top genre`) AS subquery_alias;
```
