### Window function
#### Here their is three different window functions  
1.Aggregate, 2.Ranking , 3.values

```python
%%sql
# Aggreate to perform sum,average,count,min and max on window function

select family_id,name,annual_income,
    sum(annual_income) over (partition by family_id) as total_annual_income,
    format((avg(annual_income) over (partition by family_id)),0) as avg_annual_income,
    count(*) over (partition by family_id) as total_family_members,
    max(annual_income) over (partition by family_id) as highest_income,
    min(annual_income) over (partition by family_id) as lowest_income
from family_info
```

```python
%%sql
# Ranking to perform row_number,rank,dense_rank,percent_rank,ntile

select 
    name,annual_income,
    row_number() over (order by annual_income DESC) as row_num,
    rank() over (order by annual_income DESC) as income_rank,
    dense_rank() over (order by annual_income DESC) as income_dense_rank,
    percent_rank() over (order by annual_income DESC) as income_perence_rank,
    ntile(4) over (order by annual_income DESC) as income_ntile
FROM family_info;
```

```python
%%sql
# Values to perform lag,lead,first value,last value, nth value

select name, annual_income,
    lag(annual_income) over (partition by family_id order by annual_income DESC) as annual_income_lag,
    lead(annual_income) over (partition by family_id order by annual_income DESC) as annual_income_lead,
    first_value(annual_income) over (partition by family_id order by annual_income DESC) as annual_income_first,
    last_value(annual_income) over (partition by family_id order by annual_income DESC) as annual_income_last,
    nth_value(annual_income,2) over (partition by family_id order by annual_income DESC) as annual_income_2_value
from family_info
```

### Partitioning
1. range partitioning, 2. list partitioning, 3. hash partictioning, 4. key partitioning

```python
%%sql
# range partitioning

drop table if exists sales_info;

create table sales_info(id int,sales_date date,department varchar(50),
                        amount int, primary key(id,sales_date))
partition by range(year(sales_date))(
    partition p0 values less than (2020),
    partition p1 values less than (2021),
    partition p2 values less than (2022),
    partition p3 values less than maxvalue
);

insert into sales_info(id,sales_date,department,amount) 
values
(1,'2020-01-01','hr',90000),
(2,'2020-01-04','finance',9000),
(3,'2021-01-04','marketing',40000),
(4,'2022-01-23','developer',30000),
(5,'2023-04-02','hr',40000);
```

```python
%%sql
SELECT 
    PARTITION_NAME,
    PARTITION_METHOD,
    PARTITION_EXPRESSION,
    SUBPARTITION_METHOD,
    SUBPARTITION_EXPRESSION
FROM information_schema.PARTITIONS
WHERE TABLE_SCHEMA = 'learn_sql'
  AND TABLE_NAME   = 'sales_info';
```

### Subpartitioning

```python
%%sql
# sub partitioning with range and hash partitioning

create table sales_info4(id int,sales_date date,department varchar(50),
                        amount int, primary key(id,sales_date))
partition by range(year(sales_date))
subpartition by hash(month(sales_date))
subpartitions 12(
    partition p0 values less than (2020),
    partition p1 values less than (2021),
    partition p2 values less than (2022),
    partition p3 values less than maxvalue
);

insert into sales_info4(id,sales_date,department,amount) 
values
(1,'2020-01-01','hr',90000),
(2,'2020-01-04','finance',9000),
(3,'2021-01-04','marketing',40000),
(4,'2022-01-23','developer',30000),
(5,'2023-04-02','hr',40000);
```

```python
%%sql
# 1NF 

CREATE TABLE Students (
    StudentID INT PRIMARY KEY,
    Name VARCHAR(50),
    Department VARCHAR(50)
);

CREATE TABLE Courses (
    CourseID INT PRIMARY KEY,
    CourseName VARCHAR(50)
);

CREATE TABLE StudentCourses (
    StudentID INT,
    CourseID INT,
    PRIMARY KEY (StudentID, CourseID),
    FOREIGN KEY (StudentID) REFERENCES Students(StudentID),
    FOREIGN KEY (CourseID) REFERENCES Courses(CourseID)
);

INSERT INTO Students (StudentID, Name, Department) VALUES
(1, 'John', 'Engineering'),
(2, 'Alice', 'Science');

INSERT INTO Courses (CourseID, CourseName) VALUES
(1, 'Math'),
(2, 'Science'),
(3, 'Physics'),
(4, 'Chemistry');

INSERT INTO StudentCourses (StudentID, CourseID) VALUES
(1, 1), (1, 2), (2, 3), (2, 4);
```

```python
%%sql
SELECT s.StudentID, s.Name, c.CourseName
FROM Students s
JOIN StudentCourses sc ON s.StudentID = sc.StudentID
JOIN Courses c ON sc.CourseID = c.CourseID;
```

```python
%%sql
# 2NF
drop table if exists Departments;

CREATE TABLE Departments (
    DepartmentID INT PRIMARY KEY,
    DepartmentName VARCHAR(50)
);

INSERT INTO Departments (DepartmentID, DepartmentName) VALUES
(1, 'Engineering'),
(2, 'Science');
```

```python
%%sql
SELECT s.StudentID, s.Name, d.DepartmentName
FROM Students s
JOIN Departments d ON s.DepartmentID = d.DepartmentID;

```

```python
%%sql
SELECT d.DepartmentName, COUNT(s.StudentID) AS TotalStudents
FROM Students s
JOIN Departments d ON s.DepartmentID = d.DepartmentID
GROUP BY d.DepartmentName;
```

```python
%%sql
# 3NF

CREATE TABLE Professors (
    ProfessorID INT PRIMARY KEY,
    Name VARCHAR(50),
    DepartmentID INT,
    FOREIGN KEY (DepartmentID) REFERENCES Departments(DepartmentID)
);

INSERT INTO professors 
VALUES
(1,'RAM',1),
(2,'ARUN',2);
```

# Statistical analysis
descriptive statistics, correlational analysis, regression analysis, covariance analysis, outlier detection

```python
%%sql
CREATE TABLE statistics_practice (
    id INT AUTO_INCREMENT PRIMARY KEY,
    category VARCHAR(50),
    value FLOAT);

INSERT INTO statistics_practice (category, value) VALUES
('A', 10), ('A', 15), ('A', 14), ('A', 12), ('A', 19),
('B', 30), ('B', 25), ('B', 27), ('B', 31), ('B', 29),
('C', 50), ('C', 45), ('C', 48), ('C', 46), ('C', 51);

SELECT * FROM statistics_practice;
```

Descriptive statistics

```python
%%sql
# mean
SELECT category, AVG(value) AS mean_value 
FROM statistics_practice 
GROUP BY category;
```

```python
%%sql
# Median
WITH sorted AS (
    SELECT category, value,
           ROW_NUMBER() OVER (PARTITION BY category ORDER BY value) AS row_num,
           COUNT(*) OVER (PARTITION BY category) AS total_count
    FROM statistics_practice
)
SELECT category, AVG(value) AS median_value
FROM sorted
WHERE row_num IN (FLOOR(total_count / 2), CEIL(total_count / 2))
GROUP BY category;
```

```python
%%sql
# Standard Deviation & Variance
SELECT category, 
       STDDEV_POP(value) AS stddev_pop, 
       STDDEV_SAMP(value) AS stddev_samp, 
       VAR_POP(value) AS variance_pop, 
       VAR_SAMP(value) AS variance_samp
FROM statistics_practice
GROUP BY category;
```

Regression Analysis (Linear Regression Formula

```python
%%sql

WITH stats AS (
    SELECT category, 
           (SUM(value * id) - COUNT(*) * AVG(value) * AVG(id)) / 
           (COUNT(*) * SUM(id * id) - SUM(id) * SUM(id)) AS slope,
           
           (SUM(value) * SUM(id * id) - SUM(id) * SUM(id * value)) / 
           (COUNT(*) * SUM(id * id) - SUM(id) * SUM(id)) AS intercept
    FROM statistics_practice
    GROUP BY category
)
SELECT category, slope, intercept FROM stats;
```

Outlier Detection (Using Z-score)

```python
%%sql
WITH stats AS (
    SELECT category, AVG(value) AS mean_value, STDDEV(value) AS std_dev
    FROM statistics_practice
    GROUP BY category
)
SELECT p.category, p.value,
       CASE 
           WHEN ABS(p.value - s.mean_value) > 3 * s.std_dev THEN 'Outlier'
           ELSE 'Not an Outlier'
       END AS outlier_status
FROM statistics_practice p
JOIN stats s ON p.category = s.category;
```

Correlation Calculation

```python
%%sql

SELECT category,
    (SUM(value * id) - COUNT(*) * AVG(value) * AVG(id)) /
    (SQRT((SUM(value * value) - COUNT(*) * AVG(value) * AVG(value)) * 
          (SUM(id * id) - COUNT(*) * AVG(id) * AVG(id)))
    ) AS correlation
FROM statistics_practice
GROUP BY category;
```

Covariance Calculation

```python
%%sql
SELECT category,
    (SUM(value * id) - COUNT(*) * AVG(value) * AVG(id)) / COUNT(*) AS covariance
FROM statistics_practice
GROUP BY category;
```
