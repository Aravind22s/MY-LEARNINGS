```python
# Data Types

# Numerical Types
# Integer
x = 10  
print(type(x),"x=",x)  # Output: <class 'int'>

# Float
pi = 3.14  
print(type(pi),'pi=',pi)  # Output: <class 'float'>

# Complex number
z = 3 + 4j  
print(type(z),'z=',z)  # Output: <class 'complex'>
print(z.real, z.imag)

#Sequence Typea
# String
name = "Aravind"
print(type(name),'name=',name)  # Output: <class 'str'>

# List (mutable)
fruits = ["apple", "banana", "cherry"] #storing multiple values
fruits.append("mango")  # Adding an item
print('fruites=',fruits)  # Output: ['apple', 'banana', 'cherry', 'mango']

# Tuple (immutable)
coordinates = (10, 20) # fixed data can't be changed
print(type(coordinates),'coordinates=',coordinates)  # Output: <class 'tuple'>

# Range
numbers = range(5)
print(list(numbers),'numbers=',numbers)  # Output: [0, 1, 2, 3, 4]

# Set Types
# Set (unordered, unique values)
unique_numbers = {1, 2, 3, 2} # remove duplicates
print(unique_numbers)  # Output: {1, 2, 3}

# Frozenset (immutable set)
fs = frozenset([1, 2, 3]) # fixed date
print(type(fs),'fs=',fs)  # Output: <class 'frozenset'>

# Mapping Type
# dist
person = {"name": "Aravind", "age": 25} #useful for database-like storage
print(type(person),'person name=',person["name"])  # Output: "Aravind"

# Binary Types
# bytes
b = b'Hello' #used in networking
print(b)  # Output: b'Hello'

# bytearray
ba = bytearray([65, 66, 67])
print(ba)  # Output: bytearray(b'ABC')

# memoryview
mv = memoryview(b'Hello') # used for mv = memoryview(b'Hello')
print(mv[0])  # Output: 72 (ASCII of 'H')

# None Type (Represents missing or undefined values)
result = None #Used where no value is assigned yet
if result is None:
    print("No data available.")  # Output: "No data available."
```

```python
# Immutable Types (str, tuple, frozenset)
# Immutable data types cannot be changed once defined.

# Strings are immutable
name = "Aravind"
# name[0] = "B"  # This would raise an error

# Tuples are immutable
coordinates = (10, 20)
# coordinates[0] = 5  # This would raise an error

# Frozenset is immutable
fs = frozenset([1, 2, 3])
# fs.add(4)  # This would raise an error
```

```python
# Mutable Types (list, dict, set, bytearray)
# Mutable data types can be modified after creation.

# Lists are mutable
fruits = ["apple", "banana", "cherry"]
fruits.append("mango")  
print(fruits)  # Output: ['apple', 'banana', 'cherry', 'mango']

# Dictionaries are mutable
person = {"name": "Aravind", "age": 25}
person["age"] = 26  # Modifying the dictionary
print(person)  # Output: {'name': 'Aravind', 'age': 26}

# Sets are mutable
numbers = {1, 2, 3}
numbers.add(4)
print(numbers)  # Output: {1, 2, 3, 4}

# Bytearray is mutable
ba = bytearray([65, 66, 67])
ba[0] = 68
print(ba)  # Output: bytearray(b'DBC')
```

```python
# Conversions Between Types

# Integer to Float
x = 10
y = float(x)  
print(y)  # Output: 10.0

# List to Tuple
my_list = [1, 2, 3]
my_tuple = tuple(my_list)
print(my_tuple)  # Output: (1, 2, 3)

# Dictionary Keys to List
person = {"name": "Aravind", "age": 25}
keys_list = list(person.keys()) # keys to list
values_list = list(set(person.values())) # values to lis
print(keys_list)  # Output: ['name', 'age']
print(values_list) # Output: ['Aravind',25]
```

# Variable Scope and variable Lifetime

##### Variable Scope
1. Local Scope
2. Enclosing Scope
3. Global Scope
4. Built-in Scope

```python
# Local Scope (Variables inside a function)
def local_scope_example():
    x = 5  # Local variable
    print("Inside function:", x)  # Output: Inside function: 5

local_scope_example()
# print(x)  # This would raise an error (x is not accessible outside)

# Enclosing Scope (Variables inside nested functions)
def outer_function():
    y = 10  # Enclosing variable
    
    def inner_function():
        print("Inside inner function:", y)  # Output: Inside inner function: 10
    
    inner_function()
outer_function()

# Global Scope (Variables outside functions)
global_var = "Python"

def global_scope_example():
    print("Inside function:", global_var)

global_scope_example()
print("Outside function:", global_var)  # Output: Outside function: Python
```

```python
# Built-in Scope (Predefined variables & functions)

print("Built-in Scope:", abs(-40))  # Output: Built-in Scope: 40
print("ASCII:", ascii("Python™"))   # Output: ASCII: 'Python\u2122'
print("Binary:", bin(5))            # Output: Binary: 0b101
print("Boolean:", bool(0))          # Output: Boolean: False
print("Character:", chr(65))        # Output: Character: A
print("List:", list("hello"))       # Output: List: ['h', 'e', 'l', 'l', 'o']
print("Length:", len([1, 2, 3]))     # Output: Length: 3
print("Maximum:", max([1, 10, 3]))   # Output: Maximum: 10
print("Minimum:", min([1, 10, 3]))   # Output: Minimum: 1
print("Power:", pow(2, 3))          # Output: Power: 8
print("Sorted:", sorted([5, 3, 9]))  # Output: Sorted: [3, 5, 9]
print("Sum:", sum([1, 2, 3]))        # Output: Sum: 6
print("Type:", type(42))            # Output: Type: <class 'int'>
print("Zip:", list(zip([1, 2], ['a', 'b'])))  # Output: Zip: [(1, 'a'), (2, 'b')]

print('\n')
# all()
numbers = [1, 2, 3, 4, 5]  # All are non-zero values
print('non-zero:',all(numbers))  # Output: True

bool_values = [True, True, False]  # One element is False
print('check_false:',all(bool_values))  # Output: False

mixed_list = ['', 'Hello', 'Python']  # Empty string is false
print('Is_empty:',all(mixed_list))  # Output: False

print('\n')
# any()
numbers = [0, 0, 3, 0, 0]  # Only one non-zero value
print('non-zero:',any(numbers))  # Output: True

bool_values = [False, False, False]  # No True values
print('check_false:',any(bool_values))  # Output: False

mixed_list = ['', '', 'Python']  # One non-empty string
print('Is_empty:',any(mixed_list))  # Output: True

print('\n')
#enumerate(iterable)
fruits = ["apple", "banana", "cherry"] # Assigns index to items
print('output:')
for index, fruit in enumerate(fruits):
    print(index, fruit)

print('\n')
# eval(expression) Evaluates a string as Python code
expression = "5 + 10 * 2"
result = eval(expression)  # Evaluates the string as Python expression
print('value:',result)  # Output: 25

print('\n')
# filter(function, iterable) – Filters elements based on a function
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
# Function to check if number is even
def is_even(num):
    return num % 2 == 0

filtered_numbers = list(filter(is_even, numbers))
print('filter:',filtered_numbers)  # Output: [2, 4, 6, 8]

print('\n')
# map()
numbers = [1, 2, 3, 4, 5]

# Using map() to apply the `str` function to each number
string_numbers = list(map(str, numbers))
print('Map_str:',string_numbers)  # Output: ['1', '2', '3', '4', '5']

print('\n')
# range(start, stop, step) – Generates a sequence of numbers
# Generate numbers from 1 to 10 with step 2
numbers = list(range(1, 10, 2))
print('value_range:',numbers)

print('\n')
# round(x, n) – Rounds a number to n decimal places
x = 3.14159

# Round to 2 decimal places
rounded_value = round(x, 2)
print('round_value:',rounded_value)
```

```python
# Arithmetic Operators (Used for mathematical operations)

x = 10
y = 5

print('ADD:',x+y) # Addition +
print('SUB:',x-y) # Subtraction -
print('Mul:',x*y) # Multiplication *
print('Div:',x/y) # Division (returns float) /
print('Mod:',x%y) # Modulus %
print('Power:',x**y) # Expoentiation **
print('Floor:',x//y) # Floor division //
```

```python
# Comparison Operators (Used for logical comparisons)

x = 10 
y = 5

print('Equal:',x==y) # Equal to ==
print('not_equal:',x!=y) # Not equal to !=
print('Greater than:',x>y) # Greater than >
print('Less than:',x<y) # Lessor than <
print('Greater than or equal:',x>=y) # Greater than or equal to >=
print('Less than or equal:',x<=y) # Lesser than or equal to <=
```

```python
# Logical Operators (Used in Boolean logic)

x = 7

print('AND:',x > 5 and x < 10)  # AND
print('OR:',x > 10 or x < 5)   # OR
print('NOT:',not(x>5))  # NOT
```

```python
# Assignment Operators (Used for variable assignment and modification)

x = 5
y = 10

print('y =', y)  # Assign value to variable
x += 5
print('add assign:', x)  # Add and assign +=

x -= 5
print('sub assign:', x)  # sub and assign -=

x *= 5
print('mul assign:', x)  # mul and assign *=

x /= 5
print('div assign:', x)  # div and assign /=

x %= 5
print('mod assign:', x) # mod and assign %=  

x = 5  # Reset x for exponentiation
x **= 5
print('exponentiation assign:', x)  

x = 25  # Reset x for floor division
x //= 5
print('floor assign:', x) 
```

```python
# Bitwise Operators (Used for binary operations)

x = 5  # Binary: 101
y = 3  # Binary: 011

print('AND:',x & y)  # bitwise AND &
print('OR:',x | y)  # bitwise OR |
print('XOR:',x ^ y)  # bitwise XOR ^
print('NOT',~x)  # bitwise NOT ~
print('LEFT SHIFT:',x << 2)  # bitwise left shift <<
print('RIGHT SHIFT:',x >> 2)  # bitwise right shift >>
```

```python
# Identity Operators (Used to compare memory addresses of objects)

a = [1, 2, 3]
b = a  # Assigning the same list reference

print('is:',a is b)  # IS
print('is not:',a is not b)  # NOT IS
```

```python
# Membership Operators (Used for checking presence in collections)

fruits = ["apple", "banana", "cherry"]

print("banana" in fruits)  # IN
print("grape" not in fruits)  # NOT IN
```

#### 1. Write a function that counts how many times each character appears in a given string

```python
text = "Write a function that counts how many times each character appears in a given string"

count_words = len(text.split())
print("count of words in text variable:",count_words)

def repeated_char_count(text):
    char_count = {} #initialize empty dict
    for char in text:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    return char_count

print("characted Count:",repeated_char_count(text))
```

# Loops
for,while,nested, break and continue

```python
# for loop
print('for loop')

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# while loop
print('\n')
print('while loop')

count = 1
while count <= 5:
    print(count)
    count += 1

# nested loop same as nedted loop in condition statment
print('\n')
print('nested loop')

for i in range(1, 4):  # Outer loop (rows)
    for j in range(1, 4):  # Inner loop (columns)
        print(f"{i} x {j} = {i * j}")

# break condition in for
print('\n')
print('break condition in for')

fruits = ["apple", "banana", "cherry", "mango"]
for fruit in fruits:
    if fruit == "cherry":
        break
    print(fruit)

# continue codition in for 
print('\n')
print('continue codition in for')

fruits = ["apple", "banana", "cherry", "mango"]
for fruit in fruits:
    if fruit == "banana":
        continue
    print(fruit)
```

```python
# Defining a Function
print('Defining a Function')

def greet():
    print("Hello, welcome to Python!")
greet()  # Calling the function


# Function with Parameters
print('\nFunction with Parameters')

def greet_user(name):
    print(f"Hello, {name}!")
greet_user("Aravind")  # Calling the function with an argument


# Function with Return Value
print('\nFunction with Return Value')

def add_numbers(a, b):
    return a + b  # Returns the sum of a and b  
result = add_numbers(5, 10)  # Calling the function and storing the result
print(f"Sum: {result}")  


# Function with Default Parameters
print('\nFunction with Default Parameters')   

def greet(name="Guest"):
    print(f"Hello, {name}!")  # Default name is "Guest"
greet()  # Calls with default parameter


# Lambda (Anonymous) Functions
print('/nLambda (Anonymous) Functions')

square = lambda x: x * x  # Lambda function to square a number
print(f"Square of 5: {square(5)}") 

#Function with Multiple Returns
print('/nFunction with Multiple Returns')
def get_values(a,b):
    return a+b, a-b, a*b, a/b  # Returns multiple values as a tuple
result = get_values(10, 5)  # Calling the function
print(f"Sum: {result[0]}, Difference: {result[1]}, Product: {result[2]}, Quotient: {result[3]}")

# Function with Variable Arguments (*args and **kwargs)
print('\nFunction with Variable Arguments (*args and **kwargs)')
def greet(*args, **kwargs):
    for name in args:
        print(f"Hello, {name}!")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
greet("Alice", "Bob", age=25, city="New York")  # Passing variable arguments


# Recursive Function
print('\nRecursive Function')
def factorial(n):
    if n == 0 or n == 1:
        return 1  # Base case
    else:
        return n * factorial(n - 1)  # Recursive call
result = factorial(5)  # Calling the recursive function
print(f"Factorial of 5: {result}")  
```

# High Order Function (HOF)

```python
# High Order Functions
def square(x):
    return x * x
def cube(x):
    return x * x * x
def apply_function(func, value):
    return func(value)

# Using the high order function
print("Square of 5:", apply_function(square, 5))  # Output: Square

print('\n')
# Using another high order function
def email_validator(domain):
    def is_valid_email(name):
        return f"{name}@{domain}"  # Returns a formatted email address
    return is_valid_email

# Using the high order function
gmail = email_validator("gmail.com")
yahoo = email_validator("yahoo.com")
print("Gmail:", gmail("aravind")) 
```

# Lambda function 
map, filter,reduce or sorted

```python
# lambda function
square = lambda x: x * x
print(square(5))  # Output: 25

# Lambda with Multiple Arguments
print('\nLambda with Multiple Arguments:')
add = lambda a, b: a + b
print(add(3, 7))  

# Lambda with Conditional Expression
print('\nLambda with Conditional Expression:')
max_value = lambda a, b: a if a > b else b
print(max_value(10, 20))  

# Lambda with map()
print('\nLambda with map():')
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x * x, numbers))  # Squaring each number
print(squared_numbers)  

# Lambda with filter()
print('\nLambda with filter():')
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))  # Filtering even numbers
print(even_numbers)  

# Lambda with reduce()
print('\nLambda with reduce():')
from functools import reduce
product = reduce(lambda x, y: x * y, numbers)  # Calculating product of all numbers
print(product)  

# Lambda with sorted()
print('\nLambda with sorted():')
numbers = [5, 2, 8, 1, 9]
sorted_numbers = sorted(numbers, key=lambda x: x)  # Sorting numbers
print(sorted_numbers)  
```

### closure, partial, Function composition, Callback function, Recursive function, Generator & yield

```python
# closure function
print('Closure Function:')
def outer_function(x):
    def inner_function(y):
        return x + y  # Inner function uses outer function's variable
    return inner_function
closure = outer_function(10)  # Creating a closure with x=10
print(closure(5))  

# Partial Function
print('\nPartial Function:')
from functools import partial
def greet(name, age):
    return f"Hello, {name}! You are {age} years old."
partial_greet = partial(greet, age=30)  # Fixing age to 30
print(partial_greet("Aravind"))  

# Function Annotations
print('\nFunction Annotations:')
def greet(name: str, age: int):
    return f"Hello, {name}! You are {age} years old."
print(greet("Aravind", 25))  # Calling the function with annotations

# Function Composition
print('\nFunction Composition:')
def add(x):
    return x + 2
def multiply(x):
    return x * 3
def compose_functions(f, g):
    def composed_function(x):
        return f(g(x))
    return composed_function
composed = compose_functions(add, multiply)  # Composing add and multiply functions
print(composed(5))  # Calling the composed function with 5

# Callback Functions
print('\nCallback Functions:')
def greet(name):
    return f"Hello, {name}!"
def process_data(data, callback):
    return callback(data)
result = process_data("Aravind", greet)  # Passing greet as a callback function
print(result) 

# Recursion function
print('\nRecursion Function:')
def factorial(n):
    if n == 0:  # Base case
        return 1
    else:
        return n * factorial(n - 1)
result = factorial(5)  # Calling the recursive function
print(f"Factorial of 5: {result}") 

# Generator & Yield function
print('\nGenerator & Yield Function:')
def infinite():
    n = 0
    while True:
        yield n  # Yielding the next value in the sequence
        n += 1
gen = infinite()  # Creating a generator object
print(next(gen))  

```

```python
# Inheritance 

# Single Inheritance
print("single Inheritance:")

class family:  # Parent class
    def car(self):
        print("family car is ford mustang")
class parent(family):  # Child class inheriting from family
    def son(self):
        print("family car is now your's")
p = parent()
p.car()  # Calling method from parent class
p.son()  # Calling method from child class

# Multiple Inheritance
print('\n')
print("Multiple Inheritance:")

class family:  # Parent class
    def car(self):
        print("family car is ford mustang")
    def bike(self):
        print("family bike is honda")
class parent(family):  # Child class inheriting from family
    def father(self):
        print("family car is for sales")
class childern(parent):  # Another child class inheriting from parent
    def son(self):
        print("family bike is now your's")

p = childern()
p.car()  # Calling method from parent class 
p.bike()  # Calling method from parent class
p.father()  # Calling method from parent class
p.son()  # Calling method from child class

# Hierarchical Inheritance
print('\n')
print("Hierarchical Inheritance:")

class animal:  # Parent class
    def sound(self):
        print("Animal makes sound")
class dog(animal):  # Child class inheriting from animal
    def bark(self):
        print("Dog barks")
class cat(animal):  # Another child class inheriting from animal
    def meow(self):
        print("Cat meows")
d = dog()
d.sound()  # Calling method from parent class
d.bark()  # Calling method from child class
print('\n')
c = cat()
c.sound()  # Calling method from parent class
c.meow()  # Calling method from child class

# Multiple Inheritance
print('\n')
print("Multiple Inheritance:")

class vehicle:  # Parent class
    def start(self):
        print("Vehicle started")
class two_wheeler():  # Child class inheriting from vehicle
    def ride(self):
        print("Riding two-wheeler")
class four_wheeler():  # Another child class inheriting from vehicle
    def drive(self):
        print("Driving four-wheeler")
class bike(two_wheeler, vehicle):  # Multiple inheritance
    def bike_ride(self):
        print("Riding bike")
class car(four_wheeler, vehicle):  # Multiple inheritance
    def car_drive(self):
        print("Driving car")    

b = bike()
b.start()  # Calling method from parent class
b.ride()  # Calling method from child class
b.bike_ride()  # Calling method from child class

print('\n')
c = car()
c.start()  # Calling method from parent class
c.drive()  # Calling method from child class
c.car_drive()  # Calling method from child class

# Hybrid Inheritance
print('\n')
print("Hybrid Inheritance:")

class animal:  # Parent class
    def sound(self):    
        print("Animal makes sound")
class dog(animal):  # Child class inheriting from animal
    def bark(self):
        print("Dog barks")
class cat(animal):  # Another child class inheriting from animal
    def meow(self):
        print("Cat meows")
class pet(dog, cat):  # Hybrid inheritance
    def pet_sound(self):
        print("Pet makes sound")
p = pet()
p.sound()  # Calling method from parent class
p.bark()  # Calling method from child class
p.meow()  # Calling method from another child class
```

```python
# Abstruction 
''' 
if we don't call the abstract method in child class it will raise an error
so we need to call the abstract method in child class 
without the abstract method it will not raise an error but it will not work as expected

check it out give the comment line on one of the abstractmethod it causes an error 
'''

from abc import ABC, abstractmethod # Importing ABC and abstractmethod from abc module

class webapp(ABC):  # Abstract base class
    @abstractmethod
    def login(self):
        pass  # Abstract method with no implementation

    @abstractmethod
    def logout(self):
        pass  # Abstract

    def checkout(self):
        print("Checkout process")  # Concrete method with implementation

class website1(webapp):  # Inheriting from abstract base class
    def login(self):
        print("Login to website1")  # Implementing abstract method

    def logout(self):
        print("Logout from website1")  # Implementing abstract method

class website2(webapp):  # Inheriting from abstract base class
    def login(self):    
        print("Login to website2")  # Implementing abstract method  

    def logout(self):
        print("Logout from website2")  # Implementing abstract method

wa = website1()  # Creating an instance of website1
wa.login()  # Calling login method
wa.logout()  # Calling logout method

wb = website2()  # Creating an instance of website2
wb.login()  # Calling login method
wb.logout()  # Calling logout method
```

```python
# Polymorphism
# Dynamic typeing polymorphism allows methods to be overridden in child classes
print('\nDynamic Typing Polymorphism:')
class Bird:
    def fly(self):
        print("Bird is flying.")

class Plane:
    def fly(self):
        print("Plane is flying.")

def start_flight(thing):
    thing.fly()

start_flight(Bird())   # Bird is flying
start_flight(Plane())  # Plane is flying

# Function Overriding
print('\nFunction Overriding:')

class Shape:
    def area(self):
        pass

class Circle(Shape):
    def area(self):
        return 3.14 * 5 * 5  # Area of a circle with radius 5

class Rectangle(Shape):
    def area(self):
        return 10 * 4  # Area of a rectangle

shapes = [Circle(), Rectangle()]

for shape in shapes:
    print(shape.area())
```

```python
# Access Modifiers or access specifiers
print('Access Modifiers:')
print('\nPublic: Accessible from anywhere')
class PublicClass:
    def __init__(self, name="Public Class"):
        self.name = name
    def public_method(self):
        print(f"This method is {self.name}.")
pulic= PublicClass()
pulic.public_method()  # Accessible from anywhere

# Protected: Accessible within the class and its subclasses
print('\nProtected: Accessible within the class and its subclasses')
class ProtectedClass:
    def __init__(self, name="Protected Class"):
        self._name = name  # Protected attribute
    def protected_method(self):
        print(f"This method is {self._name}.")
class SubProtectedClass(ProtectedClass):
    def sub_protected_method(self):
        print(f"Sub class method is {self._name}.")
protected = SubProtectedClass()
protected.protected_method()  # Accessible within the subclass
protected.sub_protected_method()  # Accessible within the subclass

# Private: Accessible only within the class
print('\nPrivate: Accessible only within the class')
class PrivateClass:
    def __init__(self, name="Private Class"):
        self.__name = name  # Private attribute
    def __private_method(self):
        print(f"This method is {self.__name}.")
    def access_private_method(self):
        self.__private_method()  # Accessing private method within the class
class PublicPrivateClass(PrivateClass):
    def public_method(self):
        print("Accessing private method from public method:")
        self.access_private_method()  # Accessing private method through a public method
private = PrivateClass()
private.access_private_method()  # Accessing private method through a public method
public_private = PublicPrivateClass()
public_private.public_method()  # Accessing private method through a public method
```

```python
'''Encapsulation it is a concept of wrapping data and methods into a single unit (class) 
to restrict direct access to some of the object's components. It helps in data hiding and abstraction, 
ensuring that the internal representation of an object is hidden from the outside.'''
print('Encapsulation:')

class bank:
    def __init__(self, account_number, balance):
        self.__account_number = account_number  # Private attribute
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"\nDeposited: {amount} \nNew Balance: {self.__balance}")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"\nWithdrawn: {amount}. \nNew Balance: {self.__balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")

    def get_balance(self):
        return self.__balance  # Accessing private attribute through a public method

bank_account = bank("123456789", 1000)
bank_account.deposit(500)  # Depositing money
print(f"Current Balance: {bank_account.get_balance()}")  # Getting current balance

bank_account.withdraw(200)  # Withdrawing money
print(f"Current Balance: {bank_account.get_balance()}")  # Getting current balance
```

### Instance,class,static methods

```python
# Instance,class,static methods

class MathOperations:
    def __init__(self):
        self.a = 4
        self.b = 5

    @staticmethod
    def add(a, b):
        return a + b

    @classmethod
    def multiply(cls):
        # For demonstration, use default values or create an instance
        instance = cls()
        return instance.a * instance.b

    def instance_method(self):
        return self.a - self.b

math_ops = MathOperations()
print("Static Method (Add):", MathOperations.add(4, 5))  
print("Class Method (Multiply):", MathOperations.multiply())  
print("Instance Method (Subtract):", math_ops.instance_method())
```
