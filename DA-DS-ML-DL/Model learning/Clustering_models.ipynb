### K-Means Clustering
K-Means Clustering is an unsupervised machine learning algorithm that groups similar data points into K clusters based on their similarity (distance from the cluster center).

Cluster 1

High Income

Cluster 2

Medium Income

Cluster 3

Low Income

```python
X = df
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

model = KMeans(
    n_clusters=3,
    random_state=42
)

model.fit(X_scaled)
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# inertia measures how close the data points are to their cluster centroid
inertia = model.inertia_

# Silhouette Score measures: How close a point is to its own cluster, How far it is from other clusters.
silhouette = silhouette_score(X_scaled,model.labels_)

# Measures the similarity between clusters
dbi = davies_bouldin_score(X_scaled,model.labels_)

# Measures: Distance between clusters, Compactness within clusters
chi = calinski_harabasz_score(X_scaled,model.labels_)

print("Inertia (WCSS):", inertia)
print("Silhouette Score:", silhouette)
print("Davies-Bouldin Index:", dbi)
print("Calinski-Harabasz Index:", chi)
```

### Gaussian Mixture Model (GMM)
GMM is a clustering algorithm that groups similar data points by calculating the probability of belonging to each cluster.

Soft Clustering (GMM):

Cluster A = 70%

Cluster B = 20%

Cluster C = 10%

```python
from sklearn.mixture import GaussianMixture

model = GaussianMixture(
    n_components=3,
    random_state=42
)

model.fit(X_scaled)

labels = model.predict(X_scaled)
print(labels)

probabilities = model.predict_proba(X_scaled)
print(probabilities)
```
