### Linear Regression
Linear Regression is a machine learning algorithm that predicts a numerical value by finding the best-fit straight line between input and output variables.

- House prices based on area

- Salary based on years of experience

- Sales based on advertising budget

- Student marks based on study hours

The mathematical equation of Linear Regression is:

y = β₀ + β₁x

Where:

y = Predicted output

x = Input feature

β₀ = Intercept

β₁ = Coefficient (slope)

### Decision Tree Regressor
Decision Tree Regressor is a machine learning algorithm that predicts numerical values by splitting data into smaller groups using decision rules and assigning prediction values at the leaf nodes.

Example: House Price Prediction

                Area <= 1500?

                 /        \
              Yes          No
               |            |
        Price = 25L    Area <= 2200?

                         /       \
                      Yes         No
                       |           |
                 Price = 40L   Price = 55L

If a house area is 2000 sq.ft:

2000 > 1500 → Move Right

2000 ≤ 2200 → Move Left

Predicted Price = 40 Lakhs

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(random_state=42)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

### Random Forest Resgressor
Random Forest Regressor is a supervised machine learning algorithm used to predict continuous numerical values. It works by creating multiple Decision Trees and combining their predictions to produce a more accurate and stable result.

                    Dataset
                        |
        --------------------------------
        |        |        |      |      |
      Tree1    Tree2    Tree3  Tree4  Tree5
        |        |        |      |      |
       40       42       41     39     43
        \        |        |      |     /
         \       |        |      |    /
          --------------------------
                     Average
                        |
                  Prediction
                       41

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

### K-Nearest Neighbors (KNN) Regressor
K-Nearest Neighbors (KNN) Regressor is a supervised machine learning algorithm used to predict continuous numerical values. It works by finding the K nearest data points to a new data point and predicting the average of their target values

                 New House
                  (1800)

          ┌────────┼────────┐
          │        │        │
      House A   House B   House C
      1700      1750      1850
       ₹38L      ₹40L      ₹42L

              K = 3

Prediction = (38 + 40 + 42) / 3

           = ₹40 Lakhs

```python
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor(n_neighbors=5)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

### XGBoost Regressor
XGBoost is an advanced ensemble learning algorithm that improves prediction accuracy by building multiple Decision Trees one after another, with each new tree correcting the errors of the previous trees.

              Training Data
                    │
                    ▼
               Decision Tree 1
                    │
             Predicts Output
                    │
                Calculate Error
                    │
                    ▼
               Decision Tree 2
          Learns Previous Error
                    │
                Calculate Error
                    │
                    ▼
               Decision Tree 3
          Learns Remaining Error
                    │
                    ▼
            Continue Until All Trees
                    │
                    ▼
             Final Accurate Prediction
