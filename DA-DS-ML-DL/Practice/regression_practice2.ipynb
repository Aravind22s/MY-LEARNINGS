### Feature Engineering

```python
df['price'].describe(percentiles=[0.25,0.50,0.75,0.85,0.90,1])
```

```python
# outlier removal
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.80)
IQR = Q3-Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

df = df[(df["price"]>= lower) & (df["price"] <= upper)]
df['price'].describe()
```

```python
corr = df.corr(numeric_only=True)

plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.show()
```

```python
# Feature Relationships with Target Variable
df.corr(numeric_only=True)["price"].sort_values(ascending=False)
```

```python
from scipy.stats import pointbiserialr

# Create correlation for categorical variables with price using point-biserial correlation

categorical_cols_list = df.select_dtypes(include=['object']).columns.tolist()

# Create a dictionary to store correlations
categorical_corr_dict = {}

for col in categorical_cols_list:
    # Convert categorical to numeric codes
    cat_codes = pd.Categorical(df[col]).codes
    # Calculate point-biserial correlation with price
    corr_val, _ = pointbiserialr(cat_codes, df['price'])
    categorical_corr_dict[col] = corr_val

# Convert to series and sort
categorical_corr_values = pd.Series(categorical_corr_dict).sort_values(ascending=False)

# Plot
plt.figure(figsize=(10, 6))
categorical_corr_values.plot(kind='barh')
plt.xlabel('Correlation with Price')
plt.ylabel('Categorical Features')
plt.title('Categorical Variables Correlation with Price')
plt.tight_layout()
plt.show()

print(categorical_corr_values)
```

```python
cols_to_drop = ['car_ID', 'symboling', 'doornumber', 'carbody', 'carlength', 'carwidth', 'carheight',  'enginesize', 'compressionratio',]
df = df.drop(columns=cols_to_drop)
```

```python
# using dummies to covert the categorical into numerical
df = pd.get_dummies(df, columns=['CompanyName',
 'fueltype',
 'aspiration',
 'drivewheel',
 'enginelocation',
 'enginetype',
 'cylindernumber',
 'fuelsystem'], drop_first=True)
```

```python
df
```

```python
# Min-Max Scaling - It rescales the data to a fixed range of 0 to 1.
from sklearn.preprocessing import MinMaxScaler

# Scale feature columns separately
feature_scaler = MinMaxScaler()
feature_cols = ['wheelbase', 'curbweight', 'boreratio', 'stroke', 'horsepower', 'peakrpm', 'citympg', 'highwaympg']
df[feature_cols] = feature_scaler.fit_transform(df[feature_cols])

# Scale target column separately
price_scaler = MinMaxScaler()
df['price'] = price_scaler.fit_transform(df[['price']])

df
```

```python
# assgining the target and dependent variable

y = df['price']
x = df.drop(['price'], axis=1)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=5)
X_reduced = pca.fit_transform(x)
```

```python
# split the variable for training and tesing

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size=0.2, random_state=42)

print("X_train Shape: ", X_train.shape)
print("X_test Shape: ", X_test.shape)
```

```python
from xgboost import XGBRegressor

XG_model = XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
XG_model.fit(X_train, y_train)
y_pred = XG_model.predict(X_test)
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
import numpy as np

# 1. Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# 2. Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# 3. Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# 4. R² Score
r2 = r2_score(y_test, y_pred)

# 5. Adjusted R² Score
n = X_test.shape[0]      # Number of samples
p = X_test.shape[1]      # Number of features

adjusted_r2 = 1 - ((1 - r2) * (n - 1) / (n - p - 1))

# Print Results
print(f"MAE         : {mae:.4f}")
print(f"MSE         : {mse:.4f}")
print(f"RMSE        : {rmse:.4f}")
print(f"R² Score    : {r2:.4f}")
print(f"Adjusted R² : {adjusted_r2:.4f}")
```

```python
from sklearn.neighbors import KNeighborsRegressor

KNN_model = KNeighborsRegressor(n_neighbors=5)
KNN_model.fit(X_train, y_train)
y_pred = KNN_model.predict(X_test)
```
