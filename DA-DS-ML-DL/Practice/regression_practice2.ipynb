### Data cleaning and preparation

```python
# spliting company name from CanName column

comp_name = df['CarName'].apply(lambda x: x.split(' ')[0])
df.insert(3,"CompanyName", comp_name)
df.drop(['CarName'], axis=1, inplace=True)
df.head()
```

```python
df['CompanyName'].nunique()
```

```python
df['CompanyName'].unique()
```

```python
# replacing the Company name mistake and handeling inconsisten data's
df['CompanyName'] = df['CompanyName'].str.lower()

def replace_name(a,b):
    df['CompanyName'].replace(a,b, inplace=True)
    
replace_name('maxda', 'mazda')
replace_name('porcshce', 'porsche')
replace_name('toyouta', 'toyota')
replace_name('vokswagen', 'volkswagen')
replace_name('vw', 'volkswagen')

df['CompanyName'].unique()
```
